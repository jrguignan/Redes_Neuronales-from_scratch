```python
#Codigo en keras que permite comprobar las redes echas a mano
import numpy as np
import matplotlib.pyplot as plt
import random
#import warnings
#Para quitar un warnings+-
from IPython.display import display, HTML

import tensorflow as tf
import tensorflow.keras as kr



#Se construyen los parametros de entrada y de salida
#Generacion de los datos
X =np.array(np.linspace(-4, 4, 20000))
X = X[:,np.newaxis]
#Pendiente (m), Corte en Y(b), Valores de Y sin ruido(_Y)
m , b = 0.5, 1
_Y = np.sin(b +m*X)
#Genera el ruido en los valores de _Y
Y=[]
for i in range(len(X)):
   ax = _Y[i] + ( (-1)**( (i%2)+2) )   *0.1*random.random()
   Y.append(ax)

Y=np.array(Y)


learning_rate = 0.01           # learning rate
nn = [20,3,1]  # número de neuronas por capa.


# Creamos el objeto que contendrá a nuestra red neuronal, como
# secuencia de capas.
model = kr.Sequential()

# Añadimos la capa 1
l1 = model.add(kr.layers.Dense(nn[0], activation='sigmoid'))

# Añadimos la capa 2
l2 = model.add(kr.layers.Dense(nn[1], activation='sigmoid'))

# Añadimos la capa 3
l3 = model.add(kr.layers.Dense(nn[2], activation='sigmoid'))

# Añadimos la capa 4
#l4 = model.add(kr.layers.Dense(nn[3], activation='sigmoid'))


# Compilamos el modelo, definiendo la función de coste y el optimizador.
model.compile(loss='mse', optimizer=kr.optimizers.SGD(learning_rate=0.05), metrics=['acc'])


# Y entrenamos al modelo. Los callbacks
model.fit(X, Y, epochs=100)

y = model.predict(X)


#Graficas de los puntos originales y los de la red neuronal
fig = plt.figure(figsize=(10,7))
plt.title("Grafica de Puntos Originales y los de la Red Neuronal")
plt.plot(X,_Y, color='red')
plt.scatter(X,Y)
plt.scatter(X,y)
plt.legend(["Grafica original","Puntos Entrenamiento","Puntos Red N."])
plt.show()

```
